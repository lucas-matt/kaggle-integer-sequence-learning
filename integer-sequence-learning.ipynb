# Integer Sequence Learning


**Imports**

```python
import csv
import bokeh.plotting as bplot
import numpy as np
import random
import tensorflow as tf
from tensorflow.contrib import learn
```

Parse test and training files

```python
import os
os.getcwd()

def parse(file):
    with open(file) as csvfile:
        rows = csv.reader(csvfile)
        # skip header
        next(rows)
        parsed = []
        for row in rows:
            parsed.append(
                (row[0], [int(x) for x in row[1].split(",")])
            )
        return parsed

    
train = parse('data/train.csv')
random.seed(1)
sample = random.sample(train, 25)
```

```python
def mini_plot(data):
    y = data
    x = range(len(y))
    p = bplot.figure(plot_width=180, plot_height=100)
    p.line(x, y)
    p.axis.visible = False
    return p

plots = np.array([mini_plot(data[1]) for data in sample])
grid = np.reshape(plots, (5, 5)).tolist()
pl = bplot.gridplot(grid, toolbar_location=None, logo=None)
bplot.output_notebook()
bplot.show(pl)
```
